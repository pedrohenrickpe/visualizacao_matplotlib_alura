### Fazendo as primeiras Visualizações

```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv("datasets/monitoramento_tempo.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
# Transformando a variável "data" para um objeto do tipo Datetime
df['data'] = pd.to_datetime(df['data'])
```

```python
df.info()
```

```python
plt.plot(df['data'], df['temperatura'])
```

```python
# Altera o tamanho do gráfico
plt.figure(figsize=(16, 8))
plt.title('Temperatura X Tempo')
plt.plot(df['data'], df['temperatura']);
```

```python
fig = plt.figure(figsize=(16, 8))
eixo = fig.add_axes([0, 0, 1, 1])
eixo.set_title('Temperatura X Ano', fontsize=20)
eixo.set_ylabel('Temperatura', fontsize=15)
eixo.set_xlabel('Ano', fontsize=15)
eixo.legend(['Temperatura'], loc = 'lower right', fontsize=15)
eixo.plot(df['data'], df['temperatura'], color='green');
```
